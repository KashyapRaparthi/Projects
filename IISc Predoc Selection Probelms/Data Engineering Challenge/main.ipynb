```python
import pandas as pd
import os

df = pd.DataFrame(columns=['article_id', 'text', 'category'])   #Empty df

for filename in os.listdir('BBC_articles'): # Iterating thorugh each document to extract article id, category, text
    article_id, category = filename.split('_')
    category,_ = category.split('.')
    with open(os.path.join('BBC_articles', filename), 'r') as f:
        text = f.read()

    df = df.append({'article_id': article_id, 'text': text, 'category': category}, ignore_index=True)   

df.to_csv('bbc_articles.csv', index=False) # Save the DataFrame to CSV

```

# Task 2 (Data Preprocessing for Model Training)

```python
# import nltk
# nltk.download('stopwords')  
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

df = pd.read_csv('bbc_articles.csv')  # reading the csv file

df['text'] = df['text'].str.lower() # Lowercasing 
df['text'] = df['text'].str.replace(r'[^\w\s]', '', regex=True) # remove punctuation marks

# Tokenization
df['text'] = df['text'].apply(word_tokenize)

# Removing stopwords
stop_words = stopwords.words('english')
filtered_tokens = []
for tokens in df['text']:
  filtered_tokens.append([token for token in tokens if token not in stop_words])
df['text'] = filtered_tokens
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=100) # TF-IDF vectorizer

df['text'] = df['text'].apply(lambda tokens: ' '.join(tokens))  # Converting the tokens to strings
text_features = vectorizer.fit_transform(df['text'])    # Passing the strings to the Vectorizer

features_df = pd.DataFrame(text_features.toarray(), columns=vectorizer.get_feature_names_out())
features_df['label'] = df['category']
features_df['article_ID'] = df['article_id']

features_df.to_csv('vectorized_dataset.csv', index=False) # Vectorized Dataset  
```
